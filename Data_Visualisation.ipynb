### **Data Visualisation**

```
import matplotlib.pyplot as plt
import numpy as np
```

```
#Line Plot

x = np.arange(10)
y1 = x**2
y2 = 2*x+3

print(x)
print(y1)
print(y2)
```

```
plt.style.available
```

```
plt.style.use('seaborn')
```

```
plt.plot(x,y1,color = 'red',label='Apple',marker='^')
plt.plot(x,y2,color='green',label='Kiwi',linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Prices of fruit over time')
plt.legend()
plt.show()
```

## **Scatter Plots**

```
plt.scatter(x,y1,color = 'red',label='Apple',marker='^')
plt.scatter(x,y2,color='green',label='Kiwi',linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Prices of fruit over time')
plt.legend()
plt.figure(figsize=(5,5))
plt.show()
```

## **Bargraph**

```
plt.bar?
```

```
x_cord = np.array([0,1,2])*2
plt.bar(x_cord,[10,20,15],width=0.5,label = 'current year',tick_label=['Gold','Silver','Bronze'])
plt.bar(x_cord+0.5,[20,10,12],width=0.5, label='next year')
plt.title('Metal price comparision')
plt.xlabel('Metal')
plt.ylabel('Price')
plt.xlim(-1,6)
plt.ylim(0,40)
plt.legend()
plt.show()
```

## **Pie Chart**

```
subjects = "English", "Hindi","Maths", "Physics"
weightage = [20,10,15,5]
plt.pie(weightage,labels = subjects, explode = (0,0,0.1,0))
plt.figure(figsize=(5,5))
plt.show()
```

## **Normal Distribution**

```
Xsn = np.random.randn(100)
sigma = 8
u = 70
X = np.round(Xsn*sigma + u)
X2 = np.round(Xsn*5 + 40)
print(X)
```

```
plt.hist(X, alpha=0.8)
plt.hist(X2, alpha =0.8)
plt.xlabel('Marks Range')
plt.ylabel('Prob/FreqCount of students')
plt.show()
```
